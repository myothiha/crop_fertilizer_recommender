### Merge two datasets according to their common columns. You can use other columns if you feel more appropriate.

```python
merge = True
if merge:
    df = pd.concat([df1, df2], axis=0)
else:
    df = df1

df.head(5)
```

```python
# source: https://www.indianclimate.com/

avg_humidity = {
    'Kolhapur': 71.8,
    'Solapur': 56.3,
    'Satara': 70.3,
    'Sangli': 69.8,
    'Pune': 69.3
}
```

```python
df[df['District_Name'].notna()].shape
```

```python
df.columns
```

```python
def fill_missing_humidity(item):
    district = item['District_Name']

    if district in avg_humidity.keys():
        print()
        return avg_humidity[district]
    else:
        print(item['Humidity'])
        return item['Humidity']

if merge:
    df['Humidity'] = df[['District_Name', 'Humidity']].apply(fill_missing_humidity, axis=1)
else:
    df['Humidity'] = df[['District_Name']].apply(fill_missing_humidity, axis=1)
```
