```python
#importing all the libraries
import matplotlib.pyplot as plt 
import seaborn as sns

plt.title("Pair plot for all features")
sns.pairplot(df,hue='Fertilizer')

```

```python
#Count plot for all the features
plt.figure(figsize=(15,15))

plt.title("Count plot for Nitrogen")
sns.countplot(x='Nitrogen',data=df,hue='Nitrogen')
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(10,10))

plt.title("Count plot for Phospohrous")
sns.countplot(x='Phosphorus',data=df,hue='Phosphorus',color="green")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.title("Count plot for Potassium")
sns.countplot(x='Potassium',data=df,hue='Potassium',color='red')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.title("Count plot for Fertilizer")
sns.countplot(x='Fertilizer',data=df,hue='Fertilizer',color='Yellow')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.title("Count plot for Temperature")
sns.countplot(x='Temperature',data=df,hue='Temperature',color='red')
plt.xticks(rotation=45)
plt.show()
```
