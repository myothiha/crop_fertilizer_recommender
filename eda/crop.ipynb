```python
import matplotlib.pyplot as plt 
plt.title("Relationship between the Potassium fertilizer and Crop supported")

sns.barplot(x=df['K'],y=df['label'],data=df,hue="label")

plt.xlabel("Amount of potassium")
plt.ylabel("Most yield crop using this fertilizer")

plt.show()
```

From the above plot we can see that soil with higher amount of potassium in it,is more prefable for crops like grapes and apple.
OR
From the above plot we can deduce thatfor crops like apple and grapes the amount of potassium reqired is 200 units

```python
plt.title("Between the fertlizer N and crop yeilded")

sns.barplot(x=df['N'], y=df['label'],data=df,hue="label")
plt.xticks(rotation=90)
plt.xlabel('Amount of Nitrogen required')
plt.ylabel('Particular crop yield based on the amount of Fertilizer')

plt.show()

```

From the above plot we can deduce that the soil which are rich in nitrogen content are more prefable for crops like cotton followd by muskmelon,watermelon and banana.


```python
plt.title("Between fertilizer P and crop yeild")

sns.barplot(x=df['P'],y=df['label'],hue="label",data=df)

plt.xlabel('Amount of P Fertilizer')
plt.ylabel('Crop corresponding to particular amount of P')
plt.xticks(rotation=90)
plt.show()
```

From the above plot we can deduce that the soil with more amount of Posphorous is more prefable for crops like grapes and apple.

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder()

# Fit and transform the 'label' column using OneHotEncoder
label_encoded = one_hot_encoder.fit_transform(df['label'].values.reshape(-1, 1))

# Create a DataFrame from the one-hot encoded data
one_hot_df = pd.DataFrame(label_encoded.toarray(), columns=one_hot_encoder.get_feature_names_out(['label']))

# Concatenate the one-hot encoded DataFrame with the original DataFrame
df = pd.concat([df, one_hot_df], axis=1)

# Drop the original 'label' column if needed
df = df.drop('label', axis=1)

# Print the updated DataFrame
print(df)
```

```python


# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(20,20))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Set plot title
plt.title("Correlation Heatmap")

# Show the plot
plt.show()
```
